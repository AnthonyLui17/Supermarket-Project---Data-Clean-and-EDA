Plotting income against various other columns

```python
columns=['MntWines', 'MntFruits', 'MntMeatProducts',
       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases', 'NumWebVisitsMonth', 'Total_Amount', 'Total_Purchases']
plt.rcParams['figure.figsize'] = [12, 30]
for i in range(len(columns)):
    plt.subplot(7,2,i+1)
    plt.scatter(data=df2, x='Income', y=columns[i])
    plt.xlabel('Income')
    plt.ylabel(columns[i])
plt.show()
```

Boxplot of Year of Birth, excluding outliers

```python
q1 = np.percentile(df['Year_Birth'], 25)
q3 = np.percentile(df['Year_Birth'], 75)
upper_bound = q3 + 1.5*(q3-q1)
lower_bound = q1 - 1.5*(q3-q1)
df3=df[(df['Year_Birth'] < upper_bound) & (df['Year_Birth'] > lower_bound)]
plt.rcParams['figure.figsize'] = [5, 5]
sns.boxplot(df3, x='Year_Birth')
```

Pie chart showing distribution of revenue

```python
labels = ['MntWines', 'MntMeatProducts', 'MntGoldProds',
       'MntFishProducts', 'MntSweetProducts', 'MntFruits']
total_amount=[]
for i in range(len(labels)):
    total_amount.append(df[labels[i]].sum())
plt.figure(figsize=(6,6))
plt.pie(total_amount, labels=labels, autopct='%1.0f%%')
plt.title('Distribution of Revenue')
```

Pie chart showing distribution of purchases methods
